# SIMPLE LINEAR REGRESSION

```python


X = LinRegData['Cases - cumulative total'].values.reshape(-1,1)
y = LinRegData['Deaths - cumulative total'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
print('The R-square is: ', regressor.score(X_train, y_train))
```

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
#df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten(), 'percentage error': ((ym_pred)/(ym_pred)*100)-1})
df
```

```python
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
